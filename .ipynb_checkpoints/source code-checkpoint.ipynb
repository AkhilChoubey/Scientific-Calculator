```python
from tkinter import *
from PIL import Image,ImageTk

global original_price

original_price = float(input("enter original price"))
    
        
        
class bikePriceCalculate:
    def __init__(self,root):
        self.root=root
        self.root.title("Price estimation of used bikes")
        self.root.geometry("1350x700+0+0")
        
        self.root.config(bg="cyan")
        
        #==========Background Image================#
        self.bg=ImageTk.PhotoImage(file="Images/bikeBG.jpg")
        bg=Label(self.root,image=self.bg).place(x=250,y=0,relwidth=1,relheight=1)
        
        #=========Left Image=======================#
        self.left=ImageTk.PhotoImage(file="Images/frontBike.png")
        left=Label(self.root,image=self.left).place(x=80,y=100,width=400,height=500)
        
        #==========Frame for calculation=============#
        frame1=Frame(self.root,bg="white")
        frame1.place(x=480,y=100,width=700,height=500)
        
        def calculation():
            e=int(e_year.get())
            years= 2020-e
            year_percentage=float(5*years)
            kms= int(e_Km_driven.get())
            kms_percentage=2.5*(kms/1000)
            
            calculate= original_price - 20000 
            calculate= calculate - ((calculate*year_percentage)/100)
            calculate= calculate - ((calculate*kms_percentage)/100)
            show_price.delete(0,'end')
            show_price.insert(0,calculate)    

        
        title=Label(frame1,text="PRICE CALCULATOR FOR USED BIKES",font=("TIMES NEW ROMAN",20,"bold"),bg="white",fg="Blue").place(x=50,y=50)
        
        brand_name=Label(frame1,text="Enter Brand",font=("TIMES NEW ROMAN",15,"bold"),bg="white",fg="grey").place(x=50,y=100)
        e_brand=Entry(frame1,font=("times new roman",15),bg="lightgray").place(x=50,y=130,width=250)
        
        #------------------
        model_name=Label(frame1,text="Enter Model",font=("TIMES NEW ROMAN",15,"bold"),bg="white",fg="grey").place(x=370,y=100)
        e_year=Entry(frame1,font=("times new roman",15),bg="lightgray").place(x=370,y=130,width=250)
        
        #-------------------
        year_name=Label(frame1,text="Enter Year",font=("TIMES NEW ROMAN",15,"bold"),bg="white",fg="grey").place(x=50,y=170)
        e_year=Entry(frame1,font=("times new roman",15),bg="lightgray")
        e_year.place(x=50,y=200,width=250)
        
        #-------------------
        trim_name=Label(frame1,text="Enter Trim",font=("TIMES NEW ROMAN",15,"bold"),bg="white",fg="grey").place(x=370,y=170)
        cmb_trim=Entry(frame1,font=("times new roman",15),bg="lightgray").place(x=370,y=200,width=250)
        
        #-------------------
        Km_driven=Label(frame1,text="KM Driven",font=("TIMES NEW ROMAN",15,"bold"),bg="white",fg="grey").place(x=50,y=240)
        e_Km_driven=Entry(frame1,font=("times new roman",15),bg="lightgray")
        e_Km_driven.place(x=50,y=270,width=250)
        
        #-------------------
        self.btn_calculate=ImageTk.PhotoImage(file="Images/calculatePrice.png")
        btn=Button(frame1,image=self.btn_calculate, command=calculation).place(x=385,y=260,height=35)
        
        #-------------------
        final_price=Label(frame1,text="Price is",font=("TIMES NEW ROMAN",15,"bold"),bg="white",fg="grey").place(x=50,y=310)
        show_price=Entry(frame1,font=("times new roman",15),bg="lightgray")
        show_price.place(x=50,y=340,width=250)
        
        #------------------
        self.btn_QUIT=ImageTk.PhotoImage(file="Images/quit.png")
        QUIT_btn=Button(frame1,image=self.btn_QUIT, command=root.destroy).place(x=385,y=320,height=35)
        
 
        
    
root = Tk()
obj = bikePriceCalculate(root)
root.mainloop()
```

```python
#Formula for calculating price of used bikes-

#((((original Price - 20,000)-5xno. of years %)-(2.5x(no. of km driven/1000))%))
```
